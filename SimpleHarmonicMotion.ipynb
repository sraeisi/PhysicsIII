```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-talk')
```

```
t_max = 10
dt = .01
t_list= np.arange(0, t_max, dt )
```

```
def Xoft(omega, phi, A):
    return A * np.cos( omega *t_list + phi )

def Voft(omega, phi, A):
    return -omega * A * np.sin( omega*t_list + phi )

def Aoft(omega, phi, A):
    return - (omega**2) * A * np.cos( omega *t_list + phi )
```

```
freq = .2
omega = 2*np.pi*freq
phi_0 = 0#3*np.pi/2
A = 1
plt.plot(t_list, Xoft( omega, phi_0 , A ), label = phi_0/np.pi )
freq = .2
phi_0 = 0# np.pi/2
A = 1.5
plt.plot(t_list, Xoft( omega, phi_0 , A ), label = phi_0/np.pi )
plt.legend(loc=0)
```

## Position, Velocity, Acceleration

```
def Xoft(omega, phi, A):
    return A * np.cos( omega *t_list + phi )

def Voft(omega, phi, A):
    return -omega * A * np.sin( omega*t_list + phi )

def Aoft(omega, phi, A):
    return - (omega**2) * A * np.cos( omega *t_list + phi )
```

```
phi_0 = 0
A = 1 #m
omega = 1  #  2*np.pi * freq
m = 1 #kg
k = m*omega**2
```

```
x_list = Xoft( omega, phi_0 , A )
v_list = Voft( omega, phi_0 , A )
a_list = Aoft( omega, phi_0 , A )
```

```
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, xlabel = 'Time', title = 'Oscillation in Time')
plt.plot(t_list, x_list, label = 'position')
plt.plot(t_list, v_list, label = 'velocity')
plt.plot(t_list, a_list, label = 'acceleration')
plt.legend(loc=0)
plt.show()
```

## Energy

```
KineticE_list = 1/2*m*v_list**2
PotentialE_list = 1/2*k*x_list**2
```

```
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, xlabel = 'Time', title = 'Energy in Time')
plt.plot(t_list, KineticE_list, label = 'Kinetic')
plt.plot(t_list, PotentialE_list, label = 'Potential')
plt.plot(t_list, PotentialE_list+ KineticE_list, label = 'Total')
plt.legend(loc=0)
plt.show()
```

# Damping

```
def Xoft(omega, phi, A, b):
    return A * np.exp(-b*t_list/2/m) * np.cos( omega *t_list + phi )

def Voft(omega, phi, A, b):
    return -omega * A * np.exp(-b*t_list/2/m)* np.sin( omega*t_list + phi )

def Aoft(omega, phi, A, b):
    return - (omega**2) * A * np.exp(-b*t_list/2/m)* np.cos( omega *t_list + phi )
```

```
t_max = 20
dt = .01
t_list= np.arange(0, t_max, dt )

phi_0 = 0
A = 1 #m
omega = 1  #  2*np.pi * freq
m = 1 #kg
k = m*omega**2
b = .5 # Dissipation: f = -b v

x_list = Xoft( omega, phi_0 , A, .5 )
v_list = Voft( omega, phi_0 , A, .5 )
a_list = Aoft( omega, phi_0 , A, .5 )
```

## Position, Velocity, Acceleration

```
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, xlabel = 'Time', title = 'Oscillation in Time')
plt.plot(t_list, x_list, label = 'position')
plt.plot(t_list, v_list, label = 'velocity')
plt.plot(t_list, a_list, label = 'acceleration')
plt.legend(loc=0)
plt.show()
```

# Energy

```
KineticE_list = 1/2*m*v_list**2
PotentialE_list = 1/2*k*x_list**2
```

```
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, xlabel = 'Time', title = 'Energy in Time')
plt.plot(t_list, KineticE_list, label = 'Kinetic')
plt.plot(t_list, PotentialE_list, label = 'Potential')
plt.plot(t_list, PotentialE_list+ KineticE_list, label = 'Total')
plt.legend(loc=0)
plt.show()
```

# Phase space

```
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, xlabel = 'Position', ylabel='Velocity',  title = 'Phase Space')
plt.plot(x_list, v_list)
plt.ylim(-1,1)
plt.xlim(-1,1)
plt.show()
```
