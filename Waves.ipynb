```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-talk')

from matplotlib.widgets import Slider, Button, RadioButtons
from ipywidgets import interact, fixed, IntSlider

```

```
t_max = 10
dt = .01
t_list= np.arange(0, t_max, dt )

x_max = 10
dx = .01
x_list = np.arange(0, x_max, dx )
```

```
def Yoft(omega, k, A, phi):
  def f(x,t):
    return A * np.sin( k*x - omega *t + phi )
  return f
```

## Local Oscillation

```
omega = 1 #Angular frequency
k = 2*np.pi/(10/3)     # Wave Number
phi_0 = 0 #Initial Phase
A = 1     # Amplitude
wave = Yoft(omega, k, A, phi_0)
```

```
wave(10,0)
```

```
x = 0.8
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, xlabel = 'Time', title = 'Oscillation in Time')
plt.plot(t_list, wave(x, t_list), label = 'position')
plt.show()
```

## Snapshopt

```
t = 0
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, xlabel = 'Position', title = 'Oscillation in Position')
plt.plot(t_list, wave(x_list, t))
plt.show()
```

# Superposition

```
omega = 1 #Angular frequency
k = 2*np.pi/(10/3)     # Wave Number
phi_1 = 0 #Initial Phase
phi_2 = 2/3*np.pi #Initial Phase
A = 1     # Amplitude
wave1 = Yoft(omega, k, A, phi_1)
wave2 = Yoft(omega, k, A, phi_2)

t=0


####Plot
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, xlabel = 'Position', title = f'A snapshot of the string at t = {t}')
plt.plot(x_list, wave1(x_list, t), label = 'Wave 1', alpha = .5, c= 'red')
plt.plot(x_list, wave2(x_list, t), label = 'Wave 2', alpha = .5, c= 'blue')
plt.plot(x_list, wave1(x_list, t)+wave2(x_list, t), label = 'superposition', c= 'green')
plt.vlines(10/3/4, -1, 1)
plt.legend(loc=0)
```

## standing Waves
